### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
Sex = pd.get_dummies(data['sex'], drop_first = False)
# print(Sex.head())
Embarked = pd.get_dummies(data['embarked'], drop_first = True)
Pclass = pd.get_dummies(data['pclass'], drop_first = True)
new_data = pd.concat([data,Sex,Embarked,Pclass], axis = 1)
new_data.drop(['sex','embarked','pclass'], axis = 1, inplace = True)
new_data.head()
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
new_data['age'].fillna(np.mean(new_data['age']), inplace = True)
new_data['fare'].fillna(np.mean(new_data['fare']), inplace = True)
print(new_data.shape)
````

###Normalization


---



1. Normalize the columns which have numerical data



---



The function below implements the min-max normalization

````
def normalise(dataset):
    copy = dataset
    for name in dataset.columns:
        if not isinstance(dataset[name][0],str):
            copy[name] = dataset[name] - dataset[name].min()
            copy[name] /= dataset[name].max() - dataset[name].min()
    return copy      
````

````
norm_data = normalise(new_data)
norm_data.head()
````

In this part, you will work with the library ```matplotlib```. Import the ```pyplot``` from ```matplotlib``` with an alias name ```plt```.

````
from matplotlib import pyplot as plt  
from numpy import arange 
import math
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
plt.rcParams["figure.figsize"] = (10/2.4,10/2.4) # convert inches to cm
X = [i for i in arange(-3 * math.pi, 3 * math.pi, 0.001)]
Y = [math.sin(i) for i in arange(-3 * math.pi, 3 * math.pi, 0.001)]
plt.plot(X,Y)
plt.xlabel("Time(t)")
plt.ylabel(r"$\sin (t)$")
plt.title(r"$y = \sin x$")
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
X = [i for i in arange(-2 * math.pi, 2 * math.pi, 0.001)]
Y1 = [math.sin(i) for i in arange(-2 * math.pi, 2 * math.pi, 0.001)]
Y2 = [math.cos(i) for i in arange(-2 * math.pi, 2 * math.pi, 0.001)]
Y3 = [0 for i in arange(-2 * math.pi, 2 * math.pi, 0.001)]
plt.plot(X, Y1, 'r', label = r"$\sin (x)$")
plt.plot(X, Y2, 'g', label = r"$\cos (x)$")
plt.plot(X, Y3, 'b')
plt.legend()
plt.title(r"Trigonometric Functions")
````
